Importing Libraries

```
import pandas as pd
import numpy as np
```

Reading Data

```
#Source: https://ourworldindata.org/
covid=pd.read_csv("/content/owid-covid-data.csv")
covid
```

```
#Selecting the necssaray features and values only for World.
feature_name=['location', 'date', 'new_cases', 'new_deaths','new_cases_per_million','new_deaths_per_million']
complete=covid[feature_name]
world_data=complete.loc[complete.location== 'World']

#Making date in workable form
world_data[['dates', 'month', 'year' ]]=world_data.date.str.split('/', expand=True)
world_data
```

```
#checking for missing values
world_data.isna().sum()
```

```
#removing rows containing NA values
world_data=world_data.dropna()
world_data=world_data.drop(['date','year'], axis=1)
world_data
```

```
#Rearranging the columns
world_data=world_data[['location', 'dates', 'month', 'new_cases_per_million', 'new_deaths', 'new_deaths_per_million', 'new_cases']]
world_data
```

```
#changing the precision of the columns to 2 so that less computation resources are used
w=[x for x in world_data.columns if x!= 'location']
for i in w:
  world_data[i]=world_data[i].astype(float).round(2)

world_data
```

```
world_data.to_csv(r'/content/WORLD_DATA_TRANSFORMED.csv', index=False) #storing data in CSV file 
```
