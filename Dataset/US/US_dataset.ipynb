Libraries

```
import pandas as pd
import numpy as np
import glob
```

Reading Data

```
#source: https://github.com/CSSEGISandData/COVID-19
dirt='/content/'
files = glob.glob(f'{dirt}*.csv')
new_us = None

for i, f in enumerate (files):
    if i == 0:
        new_us = pd.read_csv(f)
        new_us['date'] = f
    else:
        tmp = pd.read_csv(f)
        tmp['date'] = f
        new_us = new_us.append(tmp)
new_us
```

Cleaning Data

```
#dividing the date into three columns
new_us[['month', 'dates', 'year' ]]=new_us['date'].str.split('-', expand=True)

#Checking for missing value
print(new_us.shape)
print(new_us.isna().sum())
print(new_us.isin([0.0]).sum())
new_us.describe()
```

```
#Selecting the necessary Columns
us=new_us[['Province_State', 'Lat', 'Long_', 'Confirmed',  'Incident_Rate', 'People_Tested', 'month', 'dates'] ] #incidence=daily new COVID cases for every 100,000 people
us[['dir', 'dir2', 'months' ]]=us['month'].str.split('/', expand=True)
h2=us.drop(['dir','dir2', 'month'], axis=1)
h2
```

```
states=['Alabama', 'Alaska', 'Arizona', 'Arkansas','California', 'Colorado', 'Connecticut', 'Delaware','Florida', 'Georgia','Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland','Massachusetts', 'Michigan', 'Minnesota', 'Mississippi','Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire','New Jersey', 'New Mexico', 'New York', 'North Carolina','North Dakota', 'Ohio', 'Oklahoma','Oregon', 'Pennsylvania', 'Rhode Island','South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah','Vermont','Virginia', 'Washington','West Virginia', 'Wisconsin', 'Wyoming']
print(len(states))
```

```
# Extra rows are removed
h3=h2.loc[h2['Province_State'].isin(states)]
h3
```

Transformation of Data

```
#creating new two columns for number of cases and tests for each day from total Confirmed cases and total people tested
final_data=pd.DataFrame()
for i in states:
  
  #select each country and sort it by month and then by date
  result=h5[h5.Province_State==i].sort_values(by=['months', 'dates']) 

  #subtract next row with previous row and assign values to new columns
  result[['New_cases_per_day', 'New_tests_per_day']]=result[['Confirmed', 'People_Tested']].diff()
  final_data=final_data.append(result)

final_data
```

```
#removing NA values and extra columns
final_data =final_data.dropna().drop(['Confirmed', 'People_Tested'], axis=1)
final_data
```

```
#transforming data so that when model trains on it, it predicts the new cases for country for particular month
final_data2=pd.DataFrame()
temp=[]
for i in states:
  for j in final_data.months.unique():
    
    #calculate sum of new_cases and new_tests (separately, column-wise) for each month
    su=final_data[(final_data.Province_State==i) & (final_data.months==j)].sum() 
    o= su[['New_tests_per_day',	'New_cases_per_day']]

    #calulate mean of Inicident_rate for each month for each country
    av=final_data[(final_data.Province_State==i) & (final_data.months==j)].mean()
    r=av[['Lat','Long_','Incident_Rate']]

    q=pd.Series([i,j], index=['State', 'Month'])

    #combine all the features to form a dataframe
    temp=pd.concat([q,r,o])
    final_data2=final_data2.append(pd.DataFrame(temp).T)

final_data2
```

```
#changing the precision of the columns to 2 so that less computation resources are used
w=[x for x in final_data2.columns if x!= 'State']
for i in w:
  final_data2[i]=final_data2[i].astype(float).round(2)

final_data2
```

```
final_data2.to_csv(r'/content/US_DATA_REFINED.csv', index=False) #storing data in CSV file 
```
